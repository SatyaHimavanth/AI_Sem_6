## BFS  (Breadth First Search)

```python
def Append(l1, l2):
    for i in l2:
      if i not in l1:
          l1.append(i)
    return l1
```

```python
def BFS(g, root, goal):
    open = [root]
    closed = []
    while(goal not in closed):
        x = open.pop(0)
        if x in g.keys():
            open = Append(open, g[x])
        closed.append(x)
        print(open, closed)
```

```python
BFS(graph, 'A', 'G')
```

```python
def sort_list(list1, list2):
    zipped_pairs = zip(list2, list1)
    z = [x for _, x in sorted(zipped_pairs)]
    return z
```
